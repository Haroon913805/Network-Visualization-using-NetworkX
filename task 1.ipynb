```python
# Task 1: Network Visualization using NetworkX
# ==============================

import networkx as nx
import matplotlib.pyplot as plt

# Function to generate Erds-Rnyi graph
def generate_er_network(n, p):
    return nx.erdos_renyi_graph(n, p)

# Generate networks
G1 = generate_er_network(50, 0.05)
G2 = generate_er_network(50, 0.1)

# Plotting function
def plot_graph(G, title, ax):
    pos = nx.spring_layout(G, seed=42)
    nx.draw_networkx_nodes(G, pos, ax=ax, node_size=200, node_color='skyblue')
    nx.draw_networkx_edges(G, pos, ax=ax, width=1.5, edge_color='gray')
    nx.draw_networkx_labels(G, pos, ax=ax, font_size=7)
    ax.set_title(title)
    ax.axis('off')

# Plot both graphs
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
plot_graph(G1, 'Erds-Rnyi Graph (p=0.05)', ax1)
plot_graph(G2, 'Erds-Rnyi Graph (p=0.1)', ax2)
plt.tight_layout()
plt.show()
```
